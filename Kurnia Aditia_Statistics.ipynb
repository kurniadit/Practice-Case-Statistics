### Scenario
The data-set contains aggregate individual statistics for 67 NBA seasons. from basic box-score attributes such as points, assists, rebounds etc., to more advanced money-ball like features such as Value Over Replacement. 

### Pre-processing
In this task not all data will be used, only data in 2017. So it is necessary to do filtering at the beginning. Besides that there are some players who make team transfers in the NBA transfer market so that there is duplication of player data. Therefore you can use the df.drop_duplicates() syntax to solve this to produce the same output as the trainer. Delete columns that have as many missing values as the entire row of data. Then you can do additional preprocessing if needed or you can immediately process the data. 

```python
import pandas as pd #import library
url='https://raw.githubusercontent.com/Syukrondzeko/Fellowship/main/Seasons_Stats.csv'
df=pd.read_csv(url,index_col=0) #import data
```

```python
df2017=df[df['Year']==2017] # Filter data NBA Season Year on 2017
df2017.head()
```

```python
df2017.drop_duplicates('Player',keep='first',inplace=True)# drop duplicates players name and keep the first row
missing_value=df2017.isnull().sum() # check missing value
missing_value[missing_value>0]
```

```python
# Because variable blanl and blank2 have missing value as the entire row of data then both of them will be drop
df2017.drop(['blanl','blank2'],axis=1,inplace=True)
df2017.fillna(0.0,inplace=True) # impute the rest of variable with 0 value
```

```python
missing_value=df2017.isnull().sum() # check missing value after missing value treatment
missing_value[missing_value>0]
```

### Process clean data
##### Some goals of this project: 

1.	Who is the youngest and oldest player in the NBA in 2017 for each team (Tm) ?

```python
young=df2017.loc[df2017.groupby('Tm')['Age'].idxmin()]
young[['Player','Tm','Age']].sort_values('Age').reset_index(drop=True)#youngest player each team
```

```python
old=df2017.loc[df2017.groupby('Tm')['Age'].idxmax()]
old[['Player','Tm','Age']].sort_values('Age',ascending=False).reset_index(drop=True) #oldest player each team
```

2.	Which player has the most minutes played (MP) in each position (Pos)? 

```python
df_mp=df2017.loc[df2017.groupby('Pos')['MP'].idxmax()]
df_mp[['Player','Tm','Pos','MP']].sort_values('MP',ascending=False).reset_index(drop=True) # Highest MP
```

3.	Which team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)? 

```python
team=df2017[['Tm','TRB%','AST%','STL%','BLK%']].groupby('Tm',as_index=False).mean()
```

```python
team[team['TRB%']==team['TRB%'].max()][['Tm','TRB%']]
```

```python
team[team['AST%']==team['AST%'].max()][['Tm','AST%']]
```

```python
team[team['STL%']==team['STL%'].max()][['Tm','STL%']]
```

```python
team[team['BLK%']==team['BLK%'].max()][['Tm','BLK%']]
```

4.	Who is the best player in your opinion based on his record stats? note: you can refer to variables point (PTS), assists, rebounds, or anything else. A combination of several variables would be nice. 

```python
best_player=df2017[['Player','Pos','Tm','PTS','AST','TRB','PER','TS%','eFG%']] #get some variable to measure best player
```

```python
# To make the scale of chosen variables same, so min max scaler will be use as the additional preprocessing
from sklearn.preprocessing import MinMaxScaler #import library
scaler = MinMaxScaler()
best_player[['PTS','AST','TRB','PER','TS%','eFG%']] = scaler.fit_transform(best_player[['PTS','AST','TRB','PER','TS%','eFG%']])
```

```python
best_player['avg_measure']=best_player.mean(axis=1) #make new variable as the aggregate metric to measure overall feature
best_player[best_player['avg_measure']==best_player['avg_measure'].max()] # best player based on highest average measure
```

5.	Which team has the best average stat record of their players? Note: you can refer to points, assists, rebounds, or anything else. A combination of several variables would be nice

```python
best_team=best_player.groupby('Tm').mean() # aggregate average each variable based on Team
best_team.head()
```

```python
best_team[best_team['avg_measure']==best_team['avg_measure'].max()] ## best team based on highest average measure
```
